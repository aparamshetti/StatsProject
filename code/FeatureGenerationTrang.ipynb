```python
def uniqValues(df):
    for colName in list(df):
        nuniq = df[colName].nunique()
        print('{:<30}  {:>30}'.format(colName, nuniq))
        if nuniq <2:
            df = df.drop(columns = [colName])
    return df

train = uniqValues(train)
```
